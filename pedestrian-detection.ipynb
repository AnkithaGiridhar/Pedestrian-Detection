# 2. The Classifier
I have used two default haarcascade classifiers, namely fullbody and upperbody. upperbody was only used experimentally, and does not feature in the final outputs. I have commented out a line that uses a classifier called pedestrian.xml. This was found online on Github, but only gave good results on the video that was found on the same page, and failed on other videos. fullbody.xml gave consistent results on most images and videos.

```python
full_body_cascade = cv2.CascadeClassifier('../input/haarcascades/haarcascade_fullbody.xml')
upper_body_cascade = cv2.CascadeClassifier('../input/haarcascades/haarcascade_upperbody.xml')
#full_body_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

# 3. Pedestrian Detection on Images
This was used to test the classifier initially, as images give faster results than videos. I have commented out lines using upperbody, as this was just an experiment in case a pedestrian was not completely visible. This gave a few good results, but most detections were erroneous.
The fullbody gives good results on most images. The key 0 needs to be entered in order to continue to the video. I have not attached the images I used on my system.

```python
img = cv2.imread('../input/pictures/pedestrians2.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
bodies = full_body_cascade.detectMultiScale(gray, 1.05, 3, minSize = (80,80))
#upper_bodies = upper_body_cascade.detectMultiScale(gray, 1.05, 3, minSize = (80,80))
for (x,y,w,h) in bodies:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(247, 129, 149),2)
    #for (x2,y2,w2,h2) in upper_bodies:
        #if (x2,y2,w2,h2) == (x,y,w,h):
            #break
        #else:
            #img = cv2.rectangle(img,(x2,y2),(x2+w2,y2+h2),(247, 129, 149),2) 
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 4. Pedestrian Detection on Videos 
This is the final detection on videos. I tested this out on videos as well as my webcam, and this can be used for real-time pedestrian detection. Again, the upperbody lines ae commented out and are irrelevant, and the videos used have not been attached.
The most common error I encountered was with video_capture.read() returning None, and this was solved by cross-checking the paths of both the classifier files and the images/videos used.
After a few trials, I found that the best scaling factor was 1.05 - though it made a few erroneous detections, there were more detections than scaling factors like 1.1, 1.2 and 1.3. The minimum neighbours are set to 3, as a value of 4 or 5 results in fewer detections, while 1 or 2 neighbours result in many errors. The minimum size and maximum size parameters did not make any significant changes to my outputs. 
The waitkey value impacts the speed of the video, and a smaller value would result in the video playing faster. Press 'q' to exit.


```python
video_capture = cv2.VideoCapture('../input/videos/Rain.mp4')

cv2.namedWindow("frame")

while True:
    ret, frame = video_capture.read()
    #print(ret,frame)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    #cv2.rectangle(frame, (100, 100), (200, 200), [255, 0, 0], 2)
    bodies = full_body_cascade.detectMultiScale(frame,1.05,3)
    #upper_bodies = upper_body_cascade.detectMultiScale(gray, 1.2, 5,minSize = (120,120),maxSize = (200,200))
    for (x, y, w, h) in bodies:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (189, 94, 145), 2)
        #for (x2,y2,w2,h2) in upper_bodies:
            #if (x2,y2,w2,h2) == (x,y,w,h):
                #break
            #else: cv2.rectangle(img,(x2,y2),(x2+w2,y2+h2),(247, 129, 149),2) 
    cv2.imshow("frame", frame)

    #This breaks on 'q' key
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
